```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def MC(N,d,R0,L):
    """
    The module returns points p0 and p1
    The inputs are:
        N  : numpy array indicating the normal vector to the field of view
        d  : diameter of the field of view projection (circle)
        R0 : numpy array for the location R0 of the FOV
        L  : length of the line segment to be projected through the vehicle
    """
    a = 2*np.random.random()-1; b = 2*np.random.random()-1; c = 2*np.random.random()-1
    r  = np.array([a,b,c])
    rp = r-(r@N)*N
    rp =  rp/np.sqrt(rp@rp)
    p0 =  R0+rp*0.5*d*random.random()
    p1 = p0+L*N
    return(p0,p1)
```

```python

Norm = np.array([random.random() ,random.random(),random.random()])
r0 = np.array([random.random() ,random.random(),random.random()])
Norm = Norm/np.sqrt(Norm@Norm)
d = 10; L = 20

X = []; X1 = []; 
Y = []; Y1 = [];
Z = []; Z1 = [];
for i in range(0,500):
    a,b = MC(Norm,d,r0,L)
    X.append( a[0])
    Y.append( a[1])
    Z.append( a[2])
    X1.append(b[0])
    Y1.append(b[1])
    Z1.append(b[2])
    
    
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.scatter3D(X,Y,Z,'ro')
ax.scatter3D(X1,Y1,Z1,'r')
plt.savefig("MC.png")
```
